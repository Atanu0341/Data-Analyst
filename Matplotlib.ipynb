# Matplotlib

```python
import matplotlib.pyplot as plt
```

# Linear Graph

```python
x = [1,2,3,4]
y = [5,6,7,8]
plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3,4]
y = [5,6,7,8]
c = 'r'
plt.plot(x,y,c)
plt.show()
```

```python
x = [5,2,9,4,7]
y = [10,5,8,4,2]
plt.plot(x,y)
plt.show()
```

```python
x = [5,2,9,4,7]
y = [10,5,8,4,2]
plt.plot(x,y, marker = 'o')
plt.show()
```

```python
x = [5,2,9,4,7]
y = [10,5,8,4,2]
plt.plot(x,y)
plt.plot(x, marker = 'o')
plt.plot(y, marker = '>')
plt.show()
```

# Bar Graph

```python
x = [1,2,3,4]
y = [5,6,7,8]
plt.bar(x,y)
plt.show()
```

```python
x = [1,2,3,4]
y = [5,6,7,8]
c = ['r','g','b','y']
plt.bar(x,y, color = c)
plt.show()
```

# Scatter Plot

```python
x = [1,2,3,4,5,6]
y = [4,2,3,5,1,8]
plt.title("Scatter")
plt.xlabel("month")
plt.ylabel('number')
c = ['red','magenta','green','orange','blue','brown']
plt.scatter(x,y, color = c, s = 150)
plt.show()
```
