# Seaborn Library

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = [1,2,3,4,5,6,7,8]
var1 = [8,7,6,5,4,3,2,1]
```

Matplotlib

```python
plt.plot(var, var1)
plt.show()
```

Seaborn

```python
sns.lineplot(x = var, y = var1)
plt.show()
```

```python
df = pd.DataFrame({"var":var, "var1":var1})
df
```

```python
sns.lineplot(x = 'var', y = 'var1' , data = df)
```

```python
df1 = sns.load_dataset("penguins")
df1
```

```python
sns.lineplot(x = "bill_length_mm", y = "flipper_length_mm", data = df1)
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm", y = "flipper_length_mm", data = df1, hue = "sex")
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm", y = "flipper_length_mm", data = df1, hue = "sex", palette = "rocket_r", style = "sex", markers = ["o",">"])
plt.show()
```

```python
df1 = sns.load_dataset("penguins").head(20)
df1
```

```python
sns.lineplot(x = "bill_length_mm", y = "flipper_length_mm", data = df1, hue = "sex", palette = "rocket_r", style = "sex", markers = ["o",">"])
plt.grid()
plt.title("This is a line graph related to penguins")
plt.show()
```
